```python
import pandas as pd
import requests as req
import base64
```

```python
client_id = "REDACTED"
client_secret = "REDACTED"
# print(str(base64.b64encode((client_id + ":" + client_secret).encode())))
session = req.Session()
token_response = session.post('https://accounts.spotify.com/api/token', 
                          data={"grant_type": "client_credentials"}, 
                          headers={"Authorization": "Basic "+base64.b64encode((client_id + ":" + client_secret).encode('ascii')).decode('ascii)')}
                         )
```

```python
token = token_response.json()['access_token']
token
```

```python
track_id = "70kBsrDIvCCboQsMTFvnVl"
url = "https://api.spotify.com/v1/tracks/" + track_id
```

```python
response = req.get(url, headers={'Authorization': "Bearer " + token, "Content-Type":"application/json"})
```

```python
response.json()
```
