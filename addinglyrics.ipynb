```python
import pandas as pd
from lyricsapi import get_song, extract_lyrics
```

```python
testID = "70kBsrDIvCCboQsMTFvnVl"
song = get_song(testID)
lyrics = extract_lyrics(song)
song["error"]
```

```python
def id_to_lyrics(spotID):
    song = get_song(spotID)
    if song["error"][0]:
        return None
    return extract_lyrics(song)
```

```python
df = pd.read_csv("justSpotifyAndGenre.csv")
df
```

```python
# WARNING: THIS TAKES A WHILE
df["lyrics"] = [id_to_lyrics(spotID) for spotID in df["spotifyID"]]
```

```python
df.to_csv("withLyrics.csv", index=False)
df
```
